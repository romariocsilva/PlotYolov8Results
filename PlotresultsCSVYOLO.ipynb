```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo Excel
# Substitua 'caminho_do_arquivo.xlsx' pelo caminho do seu arquivo
arquivo_excel = 'C:/Users/Acer/Desktop/ResultadosErin.xlsx'
df = pd.read_excel(arquivo_excel)

# Verificar se as colunas 'Epoch' e 'mAP' estão presentes
if 'Epoch' in df.columns and 'mAP' in df.columns and 'precision' in df.columns:
    # Plotar o gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(df['Epoch'], df['mAP'], marker='.', linestyle='-', color='blue')
    # Configurações do gráfico
    plt.title('mAP ao longo das Épocas')
    plt.xlabel('Época')
    plt.ylabel('mAP')
    plt.grid(True)
    
    # Exibir o gráfico
    plt.show()
else:
    print("Erro: As colunas 'Epoch' e 'mAP' não foram encontradas no arquivo.")
```
